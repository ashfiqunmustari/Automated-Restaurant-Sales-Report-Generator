```python
import pandas as pd
```

```python
sales_data= pd.read_csv("https://raw.githubusercontent.com/ashfiqunmustari/Automated-Restaurant-Sales-Report-Generator/refs/heads/main/Data/Local-Restaurant-Fast-Food-Sales.csv")
```

```python
sales_data.head(10)
```

```python
sales_data['item_name'].unique()
```

```python
sales_data['date'] = sales_data['date'].str.replace('-', '/')
sales_data['date'] = pd.to_datetime(sales_data['date'], dayfirst=False, errors='raise')
sales_data['date'] = sales_data['date'].dt.strftime('%d-%m-%Y')

```

```python
sales_data.head(10)
```

```python
#Checking missing values
missing_values = sales_data.isnull().sum()
print(missing_values)
```

```python
sales_data['transaction_type'] = sales_data['transaction_type'].fillna('Unknown')
print("Mising values in 'transaction type': ",sales_data['transaction_type'].isnull().sum())
```

```python
#Checking for duplicate rows
duplicate_rows = sales_data.duplicated()
print("Number of duplicate rows:", duplicate_rows.sum())
```

```python
#Calculating total amount for each order
sales_data['total_sales_value'] = sales_data['item_price']*sales_data['quantity']
sales_data.head(10)
```

```python
#Extracting month and day of each sale
sales_data['date'] = pd.to_datetime(sales_data['date'], dayfirst=True)
sales_data['month'] = sales_data['date'].dt.month
sales_data['day_of_week'] = sales_data['date'].dt.day_name()

sales_data.head(10)
```

```python
#Categorizing orders by price range
max_sale=sales_data['total_sales_value'].max()
bin= [0, 100, 250, max_sale]
label= ['Low', 'Medium', 'High']
sales_data['price_category']= pd.cut(sales_data['total_sales_value'], bins=bin, labels=label)

sales_data[['total_sales_value', 'price_category']].head(10)
```

```python
item_summary= sales_data.groupby('item_name')['total_sales_value'].sum().reset_index().sort_values(by='total_sales_value', ascending=False)
print(item_summary)
```

```python
time_summary= sales_data.groupby('time_of_sale')['total_sales_value'].sum().reset_index()
print(time_summary)
```

```python
monthly_summary= sales_data.groupby('month')['total_sales_value'].sum().reset_index()
print(monthly_summary)
```

```python
payment_summary= sales_data.groupby('transaction_type')['total_sales_value'].sum().reset_index()
print(payment_summary)
```

```python
day_of_the_week_summary= sales_data.groupby('day_of_week')['total_sales_value'].sum().reset_index()
print(day_of_the_week_summary)
```

```python
excel_file = "Restaurant_Sales_Report.xlsx"
```

```python
from openpyxl import Workbook, load_workbook
from openpyxl.styles import Font, Alignment
from openpyxl.chart import BarChart, Reference
```

```python
with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
    item_summary.to_excel(writer, index=False, sheet_name='Sales by Item')
    time_summary.to_excel(writer, index=False, sheet_name='Sales by Time of the Day')
    day_of_the_week_summary.to_excel(writer, index=False, sheet_name='Day of the week Sales')
    monthly_summary.to_excel(writer, index=False, sheet_name='Monthly Trend')
    payment_summary.to_excel(writer, index=False, sheet_name='Payment Methods')
```

```python
wb= load_workbook(excel_file)
for sheet_name in wb.sheetnames:
    ws= wb[sheet_name]
    for row in ws[1]:
        row.font= Font(bold=True)
        row.alignment= Alignment(horizontal='center')
```

```python
ws= wb['Sales by Item']
for row in range(2, 5):
    for cell in ws[row]:
        cell.font= Font(bold=True, color="FF0000")
wb.save(excel_file)
```

```python
ws= wb['Sales by Item']
chart= BarChart()
chart.title= "Best Selling Items"
chart.x_axis.title= "Item Name"
chart.y_axis.title= "Total Sales"

data= Reference(ws, min_col=2, min_row=1, max_row=ws.max_row)
categories= Reference(ws, min_col=1, min_row=2, max_row=ws.max_row)
chart.add_data(data, titles_from_data=True)
chart.set_categories(categories)

ws.add_chart(chart, "E5")
wb.save(excel_file)
```

```python
ws= wb['Sales by Time of the Day']
chart= BarChart()
chart.title= "Sales by Time of Day"
chart.x_axis.title= "Time of the Day"
chart.y_axis.title= "Total Sales"
data1= Reference(ws, min_col=2, min_row=1, max_row=ws.max_row)
categories1= Reference(ws, min_col=1, min_row=2, max_row=ws.max_row)

chart.add_data(data1, titles_from_data=True)
chart.set_categories(categories1)

ws.add_chart(chart, "D5")
```

```python
ws= wb['Day of the week Sales']
chart= BarChart()
chart.title= "Sales by Day of the Week"
chart.x_axis.title= "Day"
chart.y_axis.title= "Total Sales"

data2= Reference(ws, min_col=2, min_row=1, max_row=ws.max_row)
categories2= Reference(ws, min_col=1, min_row=2, max_row=ws.max_row)

chart.add_data(data2, titles_from_data=True)
chart.set_categories(categories2)
ws.add_chart(chart, "D5")
```

```python
from openpyxl.chart import LineChart

ws= wb['Monthly Trend']
chart= LineChart()
chart.title= "Monthly Sales Trend"
chart.x_axis.title= "Month"
chart.y_axis.title= "Total Sales"

data3= Reference(ws, min_col=2, min_row=1, max_row=ws.max_row)
categories3= Reference(ws, min_col=1, min_row=2, max_row=ws.max_row)

chart.add_data(data3, titles_from_data=True)
chart.set_categories(categories3)
ws.add_chart(chart, "D5")
```

```python
from openpyxl.chart import PieChart
ws= wb['Payment Methods']
chart= PieChart()
chart.title= "Sales by Payment Method"
data4= Reference(ws, min_col=2, min_row=1, max_row=ws.max_row)
categories4= Reference(ws, min_col=1, min_row=2, max_row=ws.max_row)

chart.add_data(data4, titles_from_data=True)
chart.set_categories(categories4)
ws.add_chart(chart, "D5")
```

```python
excel_file = "Restaurant_Sales_Report.xlsx"
wb.save(excel_file)
```

```python
from docx import Document
from docx.shared import Inches
doc = Document()
doc.add_heading("Local Restaurant Sales Report", 0)
```

```python
total_sales= sales_data['total_sales_value'].sum()
total_orders= sales_data.shape[0]
average_order= total_sales / total_orders

doc.add_heading("Executive Summary", level=1)
doc.add_paragraph(f"Total Sales: {total_sales}")
doc.add_paragraph(f"Total Orders: {total_orders}")
doc.add_paragraph(f"Average Order Value: {average_order:.2f}")
```

```python
doc.add_heading("Top-Selling Items", level=1)
table= doc.add_table(rows=1, cols=2)
hdr_cells= table.rows[0].cells
hdr_cells[0].text= "Item Name"
hdr_cells[1].text= "Total Sales"

for _, row in item_summary.iterrows():
    cells= table.add_row().cells
    cells[0].text= str(row['item_name'])
    cells[1].text= str(row['total_sales_value'])
```

```python
import matplotlib.pyplot as plt
#Plotting sales by item names
plt.figure(figsize=(6,4))
plt.bar(item_summary['item_name'], item_summary['total_sales_value'], color='skyblue')
plt.title('Items by Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_items.png')
plt.close()


#Plotting sales by time of the day
plt.figure(figsize=(6,4))
plt.bar(time_summary['time_of_sale'], time_summary['total_sales_value'], color='orange')
plt.title('Sales by Time of Day')
plt.ylabel('Total Sales')
plt.savefig('sales_time.png')
plt.close()

#Showing payment method with a pie chart
plt.figure(figsize=(6,4))
plt.pie(payment_summary['total_sales_value'], labels=payment_summary['transaction_type'], autopct='%1.1f%%', startangle=90)
plt.title('Sales by Payment Method')
plt.savefig('payment_method.png')
plt.close()

#Plotting sales by day of the week
plt.figure(figsize=(6,4))
plt.bar(day_of_the_week_summary['day_of_week'], day_of_the_week_summary['total_sales_value'], color='green')
plt.title('Sales by Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_day.png')
plt.close()
```

```python
doc.add_page_break()
doc.add_heading('Visualizations', level=1)
doc.add_paragraph('Items by Sales:')
doc.add_picture('top_items.png', width=Inches(5))

doc.add_paragraph('Sales by Time of the Day:')
doc.add_picture('sales_time.png', width=Inches(5))

doc.add_page_break()
doc.add_paragraph('Sales by Day of the Week:')
doc.add_picture('sales_day.png', width=Inches(5), height=Inches(4))

doc.add_paragraph('Sales by Payment Method:')
doc.add_picture('payment_method.png', width=Inches(5))
```
